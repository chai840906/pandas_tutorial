```python
#1. read & write CSV

import pandas as pd

#read CSV
#encoding='gb18030', not utf-8
df = pd.read_csv('/home/chai/pandas_tutorial/000001.csv', encoding='gb18030')

df.set_index('日期', inplace=True)

print(df.head())

#write to CSV
df.to_csv('sh001.csv', encoding='gb18030')

df2 = pd.read_csv('/home/chai/pandas_tutorial/sh001.csv', encoding='gb18030', index_col=0)

print(df2.head())
```

```python
#2. rename column
df2.columns = ['Code', 'Name', 'Close', 'High', 'Low', 'Open', 'Before', 'Raise', 'Raise_Pct', 'Volume', 'Amount']
print(df2.head())

df2.to_csv('sh001-1.csv',header = False)

df3 = pd.read_csv('sh001-1.csv', names=['Date', 'Code', 'Name', 'Close', 'High', 'Low', 'Open', 'Before', 'Raise', 'Raise_Pct', 'Volume', 'Amount'], index_col=0)

print(df3.head())

df3.rename(columns={'Before':'Yesterday', 'Close':'Today'}, inplace = True)
print(df3.tail())
```
