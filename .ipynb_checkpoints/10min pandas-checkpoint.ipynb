```python
#import
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
#1. object creation

#1.1 series from list
s = pd.Series([1,3,5,np.nan,6,8])
s

#1.2 dafaframe from numpy array

dates = pd.date_range('20180101',periods=6)
dates

df = pd.DataFrame(np.random.randn(6,4), index=dates, columns=['a','b','c','d'])
df = pd.DataFrame(np.random.randn(6,4), index=dates, columns=list('ABCD'))
df

#1.3 dataframe from dict
df2 = pd.DataFrame({'A':1.,
                    'B':pd.Timestamp('20180101'),
                    'C':pd.Series(1,index=list(range(4)),dtype='float32'),
                    'D':np.array([3]*4, dtype='int32'),
                    'E':pd.Categorical(["test","train","test","train"]),
                    'F':'foo'})
df2
df2.dtypes
df
```

```python
#2. viewing data

#2.1 show head & tail
df.head()
df.tail(3)

#2.2 show index, columns & data
df.index
df.columns
df.values

#2.3 statistic summary
df.describe()

#2.3 transpose
df.T

#2.4 sorting by axis
df.sort_index(axis=1, ascending=False)
df.sort_index(axis=0, ascending=False)

#2.5 sorting by values
df.sort_values(by='B',ascending=False)
```

```python
#3 selection

#3.1 select a single column
df['A']
df.A

#3.2 select rows
df[0:2]
df['20180103':'20180105']

#3.3 select by label
df.loc[dates[1]]
df.loc[:,['A','C']]
df.loc['20180103':'20180105','A':'C']
df.loc['20180102',['B','C']]

#3.4 get a scalar value
df.loc['20180102','B']
df.loc[dates[2],'B']
#wrong
#df.at['20180102','B']
#correct
df.at[dates[2],'B']

#3.5 select by position
df.iloc[3]
#when slicing (3:5), start bound(3) is included, upperbound(5) is excluded
df.iloc[3:5,0:2]
df.iloc[[1,2,5],[0,3]]
df.iloc[1:3,:]
df.iloc[1,1]
df.iat[1,1]

#3.6 boolean indexing
df[df.A > 0]
#print(df.A >0)
df[df>0]
#print(df>0)

df2 = df.copy()
#add a new column
df2['E'] = ['one','one','two','three','four','three']
df2
#filtering
df2[df2['E'].isin(['one'])]
```

```python
#4. setting

#4.1 create a new column(series)
s1 = pd.Series([1,2,3,4,5,6],index=pd.date_range('20180102',periods=6));
s1
#add new column to df2
df2['F']=s1
#new column is aligned and trimmed by df indexes
df2

#4.2 set value by label
df2.at[dates[0],'A'] = 0
df2

#4.3 set value by position
df2.iat[0,1] = 0.0
df2

#4.4 set value using numpy array
df2.loc[:,'D'] = np.array([5]* len(df))
print(df2)

#4.5 set value with 'where'(boolean indexing)
df3 = df2.copy()
df3.loc[:,'E'] = list(np.random.randn(len(df),1)*10)
df3[df3>0] = -df3
df3
```

```python
#5. missing data
df1 = df.reindex(index=dates[0:4],columns=list(df.columns)+['G'])
df1.loc[0:2,'G'] = 1
df1.iloc[0,3]=np.nan;
df1

#5.1 drop missing data
df1.dropna(how='any')

#5.2 filling missing data
df1.fillna(value = 5)
#equivilent, but less efficient
#df1[pd.isna(df1)] = 5
df1

#5.3 get boolean mask of nan
pd.isna(df1)
```

```python
#6. operations
#6.1 stats - mean
print(df1)
#calculate mean along each column
df1.mean()
df1.mean(0)
#calculate mean along each row
df1.mean(1)


#6.2 sub
s = pd.Series([1,3,5,np.nan,6,8], index=dates)
print(s)
s = s.shift(2)
print(s)

df1.sub(s,axis='index')

#6.3 apply functions to data
#np.cumsum do cumulative sum along index direction
df1.apply(np.cumsum)

#6.4 histogramming
s = pd.Series(np.random.randint(0,7,size=10))
s
s.value_counts()

#6.5 string methods
s = pd.Series(['A','B','C','Aaba','BACa',np.nan,'CCab'])
s.str.lower()
```

```python
#7.merge
#7.1 concat
df = pd.DataFrame(np.random.randn(10,4))
pieces = [df[:3], df[3:7], df[7:]]
pieces
pd.concat(pieces)

#7.2 join
left = pd.DataFrame({'key':['foo','foo'], 'lval':[1,2]})
left
right = pd.DataFrame({'key':['foo','foo'], 'rval':[4,5]})
right
pd.merge(left, right, on='key')

left = pd.DataFrame({'key':['foo','bar'], 'lval':[1,2]})
left
right = pd.DataFrame({'key':['foo','bar'], 'rval':[4,5]})
right
pd.merge(left, right, on='key')

#7.3 append
df = pd.DataFrame(np.random.randn(8,4),columns=['A','B','C','D'])
df
s = df.iloc[3]
s
df.append(s,ignore_index = True)
```

```python
#8 Grouping
df = pd.DataFrame({'A' : ['foo', 'bar', 'foo', 'bar','foo', 'bar', 'foo', 'foo'],
                   'B' : ['one', 'one', 'two', 'three', 'two', 'two', 'one', 'three'],
                   'C' : np.random.randn(8),
                   'D' : np.random.randn(8)})
print(df)
print(df.groupby('A').sum())
df[df.A == 'bar']

df.groupby(['A','B']).sum()
```

```python
#9 Reshaping
#9.1 stack
tuples = list(zip(*[['bar', 'bar', 'baz', 'baz','foo', 'foo', 'qux', 'qux'],
                    ['one', 'two', 'one', 'two','one', 'two', 'one', 'two']]))
print(tuples)
index = pd.MultiIndex.from_tuples(tuples,names=['first','second'])
index
df = pd.DataFrame(np.random.randn(8,3),index = index, columns =['A','B','C'])
print(df)
df2= df[:4]
print(df2)

stacked = df2.stack()
print(stacked)

unstacked = stacked.unstack()
unstacked2 = unstacked.unstack()
unstacked2


#9.2 pivot table
df = pd.DataFrame({'A' : ['one', 'one', 'two', 'three'] * 3,
                   'B' : ['A', 'B', 'C'] * 4,
                   'C' : ['foo', 'foo', 'foo', 'bar', 'bar', 'bar'] * 2,
                   'D' : np.random.randn(12),
                   'E' : np.random.randn(12)})
df
pd.pivot_table(df, values='E',index=['B','C'],columns='A')
```

```python
#10 time series
rng = pd.date_range('31/7/2018',periods=100,freq='S')
ts = pd.Series(np.random.randint(0,500,len(rng)),index=rng)
#print(ts)

print(ts.resample('1Min').sum())


#10.1 time zone representation
rng = pd.date_range('31/7/2018 0:00', periods=5, freq='D')
ts = pd.Series(np.random.randn(len(rng)),rng)
print(ts) 
ts_utc = ts.tz_localize('UTC')
ts_utc

#10.2 converting to other time zone
ts_utc.tz_convert('US/Eastern')

#10.3 converting between time span representations
rng = pd.date_range('31/7/2018',periods=5, freq='M')
ts = pd.Series(np.random.randn(len(rng)), index=rng)
print(ts)

ps = ts.to_period()
print(ps)

ps.to_timestamp()

#10.4
prng = pd.period_range("2008Q1",'2018Q4',freq='Q-NOV')
ts = pd.Series(np.random.randn(len(prng)),prng)
print(ts.head())
ts.index = (prng.asfreq('M','e')+1).asfreq('H','s')+9
print(ts.head())
```

```python
#11.Categoricals
df = pd.DataFrame({"id":[1,2,3,4,5,6,7,8],"raw_grade":['a','b','a','d','d','e','a','b']})
df = df.set_index("id")
df
df["grade"]=df["raw_grade"].astype("category")
df["grade"]
#assign categories to "grade"
df["grade"].cat.categories = ["very good","good","bad","very bad"]
df
#add missing categories(set_categories)
df["grade"] = df["grade"].cat.set_categories(["very bad", "bad", "medium","good", "very good"])
df["grade"]
#show category size
df.groupby("grade").size()
```

```python
#12.plot
ts = pd.Series(np.random.randn(1000),index=pd.date_range('31/7/2018',periods=1000,freq='D'))
ts = ts.cumsum()
ts.plot()
print(ts.head())

df = pd.DataFrame(np.random.randn(1000,4), index=ts.index,columns=['A','B','C','D'])
df = df.cumsum()
df.head()
plt.figure();df.plot();plt.legend(loc='best')
```
